```python
from datetime import datetime

spotified["Date2"]=spotified["Date"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
spotified["Date2"][10]
```

```python
global_total_streams=spotified[(spotified["Region"]=="global")].groupby("Date2").sum().drop("Position",axis=1)
global_total_streams.head()
```

```python
%matplotlib inline
plt.plot(global_total_streams)
```

```python
turkey_total_streams=spotified[(spotified["Region"]=="tr")].groupby("Date2").sum().drop("Position",axis=1)
turkey_total_streams.head()
```

```python
plt.plot(turkey_total_streams)
```
