```python
c = DecisionTreeClassifier(min_samples_split=100,random_state=42)
```

```python
features = ["danceability","loudness", "valence", "energy","instrumentalness","acousticness","key","speechiness","duration_ms"]
```

```python
X_train = train[features]
y_train = train["target"]

X_test = test[features]
y_test = test["target"]
```

```python
dt = c.fit(X_train, y_train)
```

```python
def show_tree(tree,features,path):
    f = io.StringIO()
    export_graphviz(tree,out_file=f, feature_names=features)
    pydotplus.graph_from_dot_data(f.getvalue()).write_png(path)
    img = misc.imread(path)
    plt.rcParams["figure.figsize"] = (20,20)
    plt.imshow(img)
```

```python
show_tree(dt, features, 'dec_tree_01.png')
```

```python
y_pred = c.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)*100
```

```python
print("Accuracy using decision Tree", round(score,1),"% ")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
score_pairs={}
for n in range(10,110,1):
    clf = RandomForestClassifier(n_estimators = 100)
    rfc = clf.fit(X_train, y_train)
    forest_y_pred = clf.predict(X_test)
    score = accuracy_score(y_test, forest_y_pred) * 100
    rounded_score = round(score, 1)
    score_pairs[n]=rounded_score
#print("Random Forest (n_est: 100) Accuracy: {}%".format(rounded_score))
```

```python

plt.figure(figsize=(7,7))
plt.plot(score_pairs.keys(),score_pairs.values())
plt.xlabel("Number of estimators")
plt.ylabel("Accuracy")
plt.title("N vs Acc")
plt.grid(True)
plt.show()

```
